```python
import pandas as pd
import os
from os.path import join, exists

import matplotlib.pyplot as plt
```

```python
# From Dr. Meylan's telephone analysis code 3/11

def getLengthInWords(string, replacements):
    sentence_string = string.strip().split(' ')
    new_sentence_string = []
    for word in sentence_string:

        if word in replacements:
            new_sentence_string += replacements[word]
        else:
            new_sentence_string += [word]
    return(len(new_sentence_string))
```

```python
abs_folder = '/Users/nicolewong/Desktop/nlp_urop/telephone-analysis-public'

csv_names = [
    'new_model_surprisal_outputs.csv',
    'new_model_surprisal_outputs_bert.csv',
    'new_model_surprisal_outputs_gpt_m.csv',
]

col_names = [
    'gpt2_surprisals',
    'bert_surprisals',
    'gpt2_surprisals',
]


# Need to normalize this or something? 

for csv_path, col_name in zip(csv_names, col_names):
    abs_csv_path = join(abs_folder, csv_path)
    all_runs = pd.read_csv(abs_csv_path)
    col_df = all_runs[col_name]
    # You should plot the surprisals over time (so over generations), I think?
    
    # Below three lines from 3/11: The Aggregate main ipynb notebook from Dr. Meylan
    
    # normalize all language model measures by the number of words
    all_runs['length_in_words'] = [getLengthInWords(x, {})
                                   for x in all_runs['user_candidate_transcription']]
    all_runs['normalized_'+col_name] = all_runs[col_name] / all_runs['length_in_words']
    
    new_plot_df = all_runs['normalized_'+col_name]
    plt.title('normalized_'+col_name+'_no_bits')
    plt.plot(range(len(new_plot_df)), new_plot_df)
    plt.show()
```
